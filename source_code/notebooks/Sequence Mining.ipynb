```python
# Function to merge values of multiple columns into a list
def merge_columns(row):
    return [row['IND'], row['RAIN'], row['IND.1'], row['T.MAX'], row['IND.2'], row['T.MIN'], row['T.MIN.G']]

# Apply the function to each row of the DataFrame
wind['sequences'] = wind.apply(lambda row: merge_columns(row), axis=1)
```

```python
data = pd.read_csv("data/ecommerce.csv")
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data = data.drop(columns=['CustomerID', 'InvoiceDate'])
```

```python
data = data.dropna()
```

```python
# Remove records with negative number of quantities
data = data[data['Quantity'] > 0]
len(data)
```

```python
data.head()
```

```python
data.to_csv("ecommerce_cleaned.csv", index = False)
```

```python
basket = pd.pivot_table(data=data,index='InvoiceNo',columns='StockCode',values='Quantity', aggfunc='sum',fill_value=0)
basket.head()
```

```python
basket['10002'].head(10)
```

```python
def convert_into_binary(x):
    if x > 0:
        return 1
    else:
        return 0
```

```python
basket_sets = basket.applymap(convert_into_binary)
```

```python
basket_sets['10002'].head(10)
```

```python
# from apyori import apriori
from mlxtend.frequent_patterns import apriori, association_rules

frequent_itemsets = apriori(basket_sets, min_support=0.05, use_colnames=True)
```
